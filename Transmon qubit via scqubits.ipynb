```python
import numpy as np
import scqubits as scq
import matplotlib.pyplot as plt
```

# Transmon

```python
tunable_tmon = scq.TunableTransmon(
    EJmax=50.0,
    EC=0.5,
    d=0.3,
    flux=0.0,
    ng=0.5,
    ncut=30)
```

```python
print(tunable_tmon)
```

Generates a simple plot of a set of eigenvalues as a function of one parameter. The individual points correspond to the a provided array of parameter values.

```python
flux_list = np.linspace(-2,2, 200)
fig, ax = plt.subplots(ncols = 2)
tunable_tmon.plot_evals_vs_paramvals('flux', flux_list, subtract_ground=True, fig =fig,ax=ax[0])
tunable_tmon.plot_evals_vs_paramvals('flux', flux_list, ax=ax[1])

plt.tight_layout()
plt.show()
```

```python
flux_list = np.linspace(-1,0, 500)
tunable_tmon.plot_evals_vs_paramvals('flux', flux_list, subtract_ground=True)
tunable_tmon.plot_evals_vs_paramvals('flux', flux_list)

plt.tight_layout()
plt.show()
```

```python
tunable_tmon.plot_coherence_vs_paramvals('flux', flux_list,scale=1e-3)
plt.show()
```

```python
tunable_tmon.plot_dispersion_vs_paramvals('ng','flux', flux_list)
```

```python
def plot_anharmonicity(qubit,fluxes):
    anharm = np.array([])
    for f in fluxes:
        tunable_tmon.flux = f
        anharm = np.append(anharm,tunable_tmon.anharmonicity())
    plt.plot(fluxes,anharm)
```

```python
plot_anharmonicity(tunable_tmon,flux_list)
```

```python
plot_anharmonicity(tunable_tmon, np.linspace(10,30,2000))
```

```python
tunable_tmon.ng=0.5
```

```python
plot_anharmonicity(tunable_tmon, np.linspace(0,1,500))
```
